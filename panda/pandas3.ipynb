```python
#map mathod
sme = pd.read_csv('social_media_engagement1.csv')
sme.head()
```

```python
ss_plat = {"Facebook":"Fb","Instagram":"Ig",'Twitter':"X"}
sme['NewPlatform'] = sme['platform'].map(ss_plat)
sme.head()
```

```python
dummy_df=pd.get_dummies(sme['sentiment_score'],dtype=int)
dummy_df
```

```python
final_df= pd.concat((sme,dummy_df),axis=1)
final_df
```

```python
def Like(num):
    if num>0 and num<1000:
        return "low"
    elif  num>1000 and num<4000:
        return "low"
    else:
        return "High"
    
```

```python
sme["Like_Category"]= sme['likes'].apply(Like)
```
