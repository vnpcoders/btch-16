```python
#join or marge

data1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],})

data2 =pd.DataFrame( {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']})
data3 =pd.DataFrame( {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']})
```

```python
dc=pd.concat((data1,data2,data3),axis=1)
dc
```

```python
dc=pd.concat((data1,data2,data3),axis=0,ignore_index=True)
dc
```

```python
#join or marge

data1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3',"k4"],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj',"vasu"],
        'Age':[27, 24, 22, 32,23],})

data2 =pd.DataFrame( {'key': ['K0', 'K1', 'K2', 'K3',"k5"],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj',"dewas"],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons',"btech"]})
```

```python
merge=data1.merge(data2,on=['key'])
merge
```

```python
merge_outer=data1.merge(data2,on=['key'],how="outer")
merge_outer
```

```python
merge_inner=data1.merge(data2,on=['key'],how="inner")
merge_inner
```

```python
merge_right=data1.merge(data2,on=['key'],how="right")
merge_right
```

```python
merge_left=data1.merge(data2,on=['key'],how="left")
merge_left
```

```python
merge_cross=pd.merge(data1,data2,how="cross")
merge_cross
```

```python
merge_cross=pd.merge(data1,data2,how="outer",on='key',validate='one_to_one')
merge_cross
```

```python
merge_cross=pd.merge(data1,data2,how="outer",on='key',validate='one_to_many')
merge_cross
```

```python
merge_cross=pd.merge(data1,data2,how="outer",on='key',validate='many_to_many')
merge_cross
```
