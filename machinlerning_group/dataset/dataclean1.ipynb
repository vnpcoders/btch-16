```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
df=pd.read_csv('retail_store_sales.csv')
```

```python
df
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["Item"] = df["Item"].fillna(method="ffill")
df["Price Per Unit"].fillna(df['Price Per Unit'].median(),inplace=True)
df["Quantity"].fillna(df['Quantity'].median(),inplace=True)
df["Total Spent"].fillna(df['Total Spent'].median(),inplace=True)
df["Discount Applied"] = df["Discount Applied"].fillna(method="ffill")
```

```python
df.info()
```

```python
df.drop(['Transaction ID',"Customer ID",'Transaction Date'], axis=1, inplace=True)
```

```python
df
```

```python
col= df.select_dtypes(include="object").columns
from sklearn.preprocessing import LabelEncoder
leb=LabelEncoder()
for i in col:
    df[i]=leb.fit_transform(df[i])

bol_leb=LabelEncoder()
df['Discount Applied']=bol_leb.fit_transform(df['Discount Applied'])
```

```python
Q1= df['Item'].quantile(0.25)
Q3=df['Item'].quantile(0.75)
IQR=Q3-Q1
IQR
```

```python
lower =Q1-1.5*IQR
upper =Q3+1.5*IQR
```

```python
df.drop(df[df["Item"]>upper].index ,inplace=True)
df.drop(df[df["Item"]<lower].index, inplace=True)
```

```python
df
```

```python
x=df.iloc[:,:7]
y=df.iloc[:,7]
x
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
st.fit_transform(x)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
y_pred=logmodel.predict(x_test)
y_pred[:10]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy')
dt.fit(x_train,y_train)
```

```python
dt_pred=dt.predict(x_test)
```

```python
accuracy_score(dt_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(criterion="entropy", n_estimators=50)
rf.fit(x_train,y_train)
```

```python
rf_pred=rf.predict(x_test)
```

```python
accuracy_score(y_test,rf_pred)
```
