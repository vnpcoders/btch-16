```python
#  import the librseris
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
#load the data set
df=pd.read_csv("DIA_trainingset_RDKit_descriptors.csv")
df
```

```python
#drop the unneccessary colum
df.drop('SMILES',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
X=df.iloc[:,1:]
y= df.iloc[:,0]
```

```python
X
```

principal component analysis

```python
#principal component analysis

from sklearn.decomposition import PCA
pca=PCA(n_components=15)
pca_x=pca.fit_transform(X)
```

```python
pca_x
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y=train_test_split(pca_x,y,test_size=0.3)
```

```python
train_X.shape
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
log_model.fit(train_X,train_y)
```

```python
pred_y=log_model.predict(test_X)
```

```python
from sklearn import metrics
metrics.accuracy_score(pred_y,test_y)*100
```
