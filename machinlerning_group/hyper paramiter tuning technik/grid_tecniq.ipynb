```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import pickle
import warnings
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv(r'Breast Cancer Winscoin.csv')
df
```

```python
df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
x=df.drop("diagnosis",axis=1)
y=df['diagnosis']
```

```python
from sklearn.preprocessing import LabelEncoder
lab_enc=LabelEncoder()
df['diagnosis']=lab_enc.fit_transform(df['diagnosis'])
```

```python
x=df.drop("diagnosis",axis=1)
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
log_model=LogisticRegression()

# Define parameter grid
grid = {
    'C': [0.001,0.01,0.1,1, 10,100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga','sag'],
    'max_iter':[10,100,1000]
}

grid_cv=GridSearchCV(estimator=log_model,
                     param_grid=grid,
                     cv=5,scoring="accuracy")
```

```python
model=grid_cv.fit(x_train,y_train)
```

```python
model.best_estimator_
```

```python
y_pred=model.predict(x_test)
y_pred[:10]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
grid = {
    'C': [0.001,0.01,0.1,1, 10,100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga','sag'],
    'max_iter':[10,100,1000]
}
random_scv=RandomizedSearchCV(estimator=log_model,param_distributions=grid,n_iter=10,)
random_model=random_scv.fit(x_train,y_train)
random_model
```

```python
y_pred_random=random_model.predict(x_test)
```

```python
accuracy_score(y_pred_random,y_test)
```
