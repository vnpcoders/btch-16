```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url= "https://webscraper.io/test-sites/e-commerce/allinone"
```

```python
Response = requests.get(url)
Response  # check if the request was successful   
```

```python
Response.content
```

```python
soup = BeautifulSoup(Response.text,'html.parser')
#soup
```

```python
# Extract the title
name = soup.find_all('a',class_ = 'title')
name_list = []
for i in name:
    title = i.text
    name_list.append(title.strip())
print(name_list)
```

```python
name2 = [i.text.strip() for i in name]
name2
```

```python
# Extract the price

price = soup.find_all('h4',class_ = 'price float-end card-title pull-right')
price_list = []
for i in price:
    prices = i.text
    price_list.append(prices.strip())
print(price_list)
```

```python
# Price Extraction
price = soup.find_all('h4',class_ = 'price float-end card-title pull-right')
price_list = []
for i in price:
    prices = i.text
    price_list.append(prices.strip())
print(price_list)
```

```python
# Extract the description
desc = soup.find_all('p',class_ = 'description card-text')
desc_list = []
for i in desc:
    description = i.text
    desc_list.append(description.strip())
print(desc_list)
```

```python
# Extract Review

review = soup.find_all('p',class_ = 'review-count float-end')
review_list = []
for i in review:
    review_ = i.text
    review_list.append(review_.strip())
print(review_list)
```

```python
# convert this information into dataframe

df = pd.DataFrame({
    "Title":name_list,
    "Price":price_list,
    "Description":desc_list,
    "Review":review_list
})
```

```python
url='https://www.flipkart.com/mobiles/pr?sid=tyy,4io&otracker=categorytree'
Response = requests.get(url)
```

```python
Box=soup.findAll('div', class_='tUxRFH')
Box
```

```python

product_name=[]
storage_info=[]
camera_info=[]
batry_info=[]
warrenty_name=[]
price_info=[]

for i in Box:
    product=i.find('div', class_='KzDlHZ')
    storage=i.find('li', class_='J+igdf') 
    camera=i.find('li', class_='J+igdf')
    battery=i.find('li', class_='J+igdf')
    warrenty=i.find('li', class_='J+igdf')
    price=i.find('div', class_='Nx9bqj _4b5DiR')


    product_name.append(product.text.strip())
    storage_info.append(storage.text.strip())
    camera_info.append(camera.text.strip())
    batry_info.append(battery.text.strip())
    warrenty_name.append(warrenty.text.strip())
    price_info.append(price.text.strip())
print("")
print(product_name)
print(storage_info) 
print(camera_info)
print(batry_info)
print(warrenty_name)
print(price_info)
```

```python
product_name = []
storage_info = []
camera_info = []
batry_info = []
warrenty_name = []
price_info = []

for i in Box:
    # Extract product name
    product = i.find('div', class_='KzDlHZ')    
    product_name.append(product.text.strip())
    
    # Extract all li tags with class 'J+igdf'
    details = i.find_all('li', class_='J+igdf')
    storage_info.append(details[0].text.strip() if len(details) > 0 else "")
    camera_info.append(details[1].text.strip() if len(details) > 1 else "")
    batry_info.append(details[2].text.strip() if len(details) > 2 else "")
    warrenty_name.append(details[3].text.strip() if len(details) > 3 else "")

    # Extract price
    price = i.find('div', class_='Nx9bqj _4b5DiR')
    price_info.append(price.text.strip() if price else "")

print(product_name)
print(storage_info)
print(camera_info)
print(batry_info)
print(warrenty_name)
print(price_info)
```

```python
df = pd.DataFrame({
    "PRODUCT":product_name,
    "STORAGE":storage_info,
    "CAMERA":camera_info,
    "BATRY":batry_info,
    "WARRENTY":warrenty_name,
    "PRICE":price_info
})
```

```python
image_file=[]
product_Name=[]
price_Info=[]
dilivery_type=[]

for i in Box:
     # Extract product name
    image = i.find('img', class_='AvifImage__ImageWrapper-sc-1055enk-0 jYiIkD')    
    image_file.append(image.img['src'] if image and image.img else "")
    

    product = i.find('p', class_='sc-eDvSVe gQDOBc NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU')    
    product_Name.append(product.text.strip())

    price = i.find('h5', class_='sc-eDvSVe dwCrSh')    
    price_Info.append(price.text.strip())

    dilivery = i.find('div', class_='sc-iBYQkv kHbCnL NewProductCardstyled__BadgeRow-sc-6y2tys-16 eznceS NewProductCardstyled__BadgeRow-sc-6y2tys-16 eznceS')    
    dilivery_type.append(dilivery.text.strip())
print()
print(image_file)
print(product_Name)
print(price_Info)
print(dilivery_type)
```

```python
image_file = []
product_Name = []
price_Info = []
delivery_type = []

for i in Box:
    # Extract product image URL
    image = i.find('img', class_='AvifImage__ImageWrapper-sc-1055enk-0 jYiIkD')
    image_file.append(image['src'] if image else "")

    # Extract product name
    product = i.find('p', class_='sc-eDvSVe gQDOBc NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU')
    product_Name.append(product.text.strip() if product else "")

    # Extract price
    price = i.find('h5', class_='sc-eDvSVe dwCrSh')
    price_Info.append(price.text.strip() if price else "")

    # Extract delivery info
    delivery = i.find('div', class_='sc-iBYQkv kHbCnL NewProductCardstyled__BadgeRow-sc-6y2tys-16 eznceS NewProductCardstyled__BadgeRow-sc-6y2tys-16 eznceS')
    delivery_type.append(delivery.text.strip() if delivery else "")

# Debug print
print("\nImage URLs:", image_file)
print("Product Names:", product_Name)
print("Prices:", price_Info)
print("Delivery Types:", delivery_type)

# from IPython.display import Image, display

# for img_url in image_file:
#     if img_url and img_url.startswith("http"):
#         display(Image(url=img_url))

from IPython.display import display, HTML

html = "<div style='display:flex; flex-wrap:wrap; gap:15px;'>"
for img_url in image_file:
    if img_url and img_url.startswith("http"):
        html += f"""
        <div style="text-align:center;">
            <img src="{img_url}" style="max-width:200px; height:auto; border:1px solid #ccc; border-radius:8px;">
        </div>
        """
html += "</div>"

display(HTML(html))
```
