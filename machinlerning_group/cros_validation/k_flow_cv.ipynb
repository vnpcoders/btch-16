```python
for col in X.columns:
    if X[col].dtype == 'object':  
        X[col] = X[col].replace("?", np.nan)   
        X[col] = X[col].fillna(X[col].mode()[0])  
    else:
        X[col] = X[col].fillna(X[col].mode()[0])
```
