```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import pickle
import warnings
warnings.filterwarnings("ignore")

```

```python
df=pd.read_csv(r"D:\btch 16\machinlerning_group\banklon\bankloan.csv")
```

```python
df
```

```python
df.info()
```

```python
x=df.drop(columns=["Personal.Loan","ID","ZIP.Code"])
y=df['Personal.Loan']
```

```python
from sklearn.preprocessing import StandardScaler
st_scaler=StandardScaler()
st_scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=(0.3))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(criterion='entropy',n_estimators=50)
rf.fit(x_train,y_train)
```

```python
with open("rf_model.pkl","wb") as f:
    pickle.dump(rf,f)
```

```python
y_pred=rf.predict(x_test)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="entropy")
dt.fit(x_train,y_train)
```

```python
with open("dt_model.pkl","wb") as f:
    pickle.dump(dt,f)
```

```python
yd_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,yd_pred)
```
