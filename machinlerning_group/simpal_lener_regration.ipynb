```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the dataset
df = pd.read_csv("C:\\Users\\vasud\\Downloads\\Salary Data.csv")
df
```

```python
# remove unnecessary columns
# Here we are dropping the columns that are not needed for the regression analysis
new_df=df.drop(["Age","Gender","Education Level","Job Title"],axis=1)
new_df
```

```python
# Check for null values in the new DataFrame
new_df.isnull().sum()
```

```python
#filling the null values with median values
new_df["Salary"].fillna(new_df["Salary"].median(),inplace=True)
new_df["Years of Experience"].fillna(new_df["Years of Experience"].median(),inplace=True)
```

```python
# Check for null values in the new DataFrame
new_df.isnull().sum()
```

```python
# Split the data into features and target variable
#dependent variable or independent variable
x= np.array(new_df["Years of Experience"]).reshape(-1, 1)
y= np.array(new_df["Salary"])
```

```python
x
```

```python
y
```

```python
# splitting the data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
#model building
from sklearn.linear_model import LinearRegression
LinReg=LinearRegression()
LinReg.fit(x_train,y_train)
```

```python
#model pridection
y_pred=LinReg.predict(x_test)
```

```python
y_pred
```

```python
act_pred_df = pd.DataFrame({"Actual":y_test,
                            "Predicted":y_pred})
act_pred_df
```

```python
# model accuracy
LinReg.score(x_train,y_train)*100
```

```python
# model accuracy
LinReg.score(x_test,y_test)*100
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)*100
```
