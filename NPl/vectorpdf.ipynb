```python
from langchain.document_loaders import PyPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

# PDF LOder

```python
loader =PyPDFLoader(r"D:\btch 16\NPl\NIPS-2017-attention-is-all-you-need-Paper.pdf")
document=loader.load()
document
```

```python
text_splitter=CharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=20
)
docs= text_splitter.split_documents(document)
docs
```

# Embidding


```python
embedding=OllamaEmbeddings(model="mxbai-embed-large")
db= FAISS.from_documents(docs,embedding)
db
```

```python
query="what are Recurrent neural network?"
query_result=embedding.embed_query(query)
query_result
```

```python
db_result_output=db.similarity_search(query)
db_result_output
```

```python
db_result_output=db.similarity_search(query)
db_result_output[0].page_content
```

```python
doc_and_score=db.similarity_search_with_score(query)
doc_and_score
```

# saving and loading

```python
# saving and loading
db.save_local("attention_all_you_need_vs_db")
```

```python
new_db=FAISS.load_local('attention_all_you_need_vs_db',
                        embedding,
                        allow_dangerous_deserialization=True)
query="what is attention function?"
docss=new_db.similarity_search(query,k=2)
docss
```
