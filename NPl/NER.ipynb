```python
docs="""Different regions of India specialize in unique and flavorful dishes that reflect their local culture and traditions. For
instance, Indore is famous for its Indori Poha-Jalebi, a popular breakfast combination, and the iconic Vijay Chaat House. Mumbai, on the
other hand, is known for its street food delights such as Vada Pav, often called the “Indian burger,” and Bhelpuri, a tangy snack made
with puffed rice, chutneys, and spices. Similarly, Kolkata is renowned for its Kathi Rolls and Mishti Doi, Delhi for its Chole Bhature 
and Paranthas from Chandni Chowk, Hyderabad for its aromatic Biryani, and Amritsar for its rich, buttery Amritsari Kulcha and Lassi. These 
diverse regional cuisines not only highlight the vast culinary landscape of India but also showcase the deep connection between food and 
cultural identity."""
all_city=[]
corpus = nlp(docs)
for corp in corpus.ents:
    if corp.label_=="GPE":
        all_city.append(corp)
print(all_city)
print(len(all_city))
```

```python
from transformers import pipeline 
import torch
ner = pipeline("ner", model="cahya/NusaBert-ner-v1.3", 
               grouped_entities=True)
text = "Jakarta, April 2025 - Polisi mengungkap sosok teman pemberi uang palsu kepada artis Sekar Arum Widara. Sosok tersebut ternyata adalah Bayu Setio Aribowo (BS), pegawai nonaktif Garuda yang ditangkap Polsek Tanah Abang di kasus serupa."
result = ner(text)
print(result)
```

```python
from transformers import pipeline
import pickle
pipe = pipeline('text-to-speech', model = "suno/bark-small")

text = "python is a high level, general-purpose programming language"

result = pipe(text)

print(result)
```

```python
from transformers import AutoModelForTextToSpeech, AutoProcessor

# Save model weights
pipe.model.save_pretrained("text_to_voice_model", safe_serialization=False)

# If processor exists, save it
if hasattr(pipe, "processor") and pipe.processor is not None:
    pipe.processor.save_pretrained("text_to_voice_model")

# If tokenizer exists, save it
if hasattr(pipe, "tokenizer") and pipe.tokenizer is not None:
    pipe.tokenizer.save_pretrained("text_to_voice_model")
```
