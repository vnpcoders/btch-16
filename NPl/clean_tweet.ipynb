```python
import pandas as pd
```

```python
df=pd.read_csv("Tweet Sentiment Analysis.csv")
df.head()
```

```python
df['tweet']=df.iloc[:,0].str.lower()

df.head()
```

```python
import regex as re
```

```python
def clean_tweet(text):
    pattern=re.compile(r'<.*?>')
    text=pattern.sub(r'',text)
    return text
```

```python
df["tweet"]=df["tweet"].apply(clean_tweet)
```

```python
df.head()
```

```python
def remove_link(text):
    patten=re.compile(r"https?://\S+|www\.\S+")
    return patten.sub(r'',text)
```

```python
df['tweet']=df['tweet'].apply(remove_link)
```

```python
chat_words = {
    'AFAIK':'As Far As I Know',
    'AFK':'Away From Keyboard',
    'ASAP':'As Soon As Possible',
    "FYI": "For Your Information",
    "ASAP": "As Soon As Possible",
    "BRB": "Be Right Back",
    "BTW": "By The Way",
    "OMG": "Oh My God",
    "IMO": "In My Opinion",
    "LOL": "Laugh Out Loud",
    "TTYL": "Talk To You Later",
    "GTG": "Got To Go",
    "TTYT": "Talk To You Tomorrow",
    "IDK": "I Don't Know",
    "TMI": "Too Much Information",
    "IMHO": "In My Humble Opinion",
    "ICYMI": "In Case You Missed It",
    "AFAIK": "As Far As I Know",
    "BTW": "By The Way",
    "FAQ": "Frequently Asked Questions",
    "TGIF": "Thank God It's Friday",
    "FYA": "For Your Action",
    "ICYMI": "In Case You Missed It",
}
```

```python
def chat_convert(text):
    new_text=[]
    for w in text.split():
        if w.upper() in chat_words.keys():
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)  
```

```python
df['tweet']=df['tweet'].apply(chat_convert)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
stopwords.words('english')
```

```python
from nltk.tokenize import word_tokenize
def remove_stopwords(text,Language="english"):

    stop_words=set(stopwords.words("english"))
    word_token= word_tokenize(text)
    filtered_sentence=[w for w in word_token if not w in stop_words]
    return filtered_sentence
```

```python
df['tweet']=df['tweet'].apply(remove_stopwords)
```
